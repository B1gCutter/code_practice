```python
a = int(input())        # 输入一个自然数，输出所有质数因子
y = 2                   
list = []               # 创建一个空列表

while a != y:           
    if a % y == 0:      
        list.append(y)  # 被除数和y不相等时，如果y为a的因数，导入到一个列表中
        a /= y          # 新自然数变为被因数除去之后的自然数，继续循环
    else:
        y += 1          # y被除不尽时+1，继续循环
list.append(int(a))     # 最后剩的a也是因数
print(list)


```

```python
import math
a = int(input("1边"))
b = int(input("2边"))
c = int(input("3边"))
p = (a+b+c)/2
s = math.sqrt(p * (p - a) * (p - b) * (p - c))
print(s)
```

```python
import datetime

year,month,day = map(int,input().split(" "))

first_date = datetime.datetime(year,1,1)
now = datetime.datetime(year,month,day)
print((now - first_date).days + 1)
```

```python
# 找斐波那契数列中的第n项
num_position = int(input("要查找斐波那契数列的第几项呢？"))
list1 = [1,1]

for i in range(1,num_position - 1):
    if i != num_position:
        new_num = list1[-1] + list1[-2]
        list1.append(new_num)

print(list1)
print(f"第{num_position}项是{list1[-1]}")
```
