```python
# 41.按位与
print(3 & 5)        # 0011和0101 -> 0001
print(4 & 12)       # 0100和1100 -> 0100(4)
```

```python
# 42.按位或(参加运算的两个数只要两个数中的一个为1，结果就为1)
print(3 | 7)    
```

```python
# 43.按位异或(相同为0，相异为1)
print(3 ^ 9)    # 1100 = 10
```

```python
# 44.按位取反:0 -> 1, 1 -> 0
# ~x = -(x + 1)
print(~3)
```
