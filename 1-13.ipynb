```python
# 一百以内偶数
list = []               #创建列表
for i in range(1, 100): #1-100范围内循环，被2整除的数放入list
    if i % 2 == 0:
        list.append(i)
print(list)             #打印列表
```

```python
# 一百以内奇数
list = []               #创建列表
for i in range(1, 100): #1-100范围内循环，被3整除的数放入list
    if i % 2 == 1:
        list.append(i)
print(list)             #打印列表

list = []               #创建列表
for i in range(1, 100, 2): #1-100范围内循环，被3整除的数放入list
        list.append(i)
print(list)             #打印列表
```

```python
#判断输入是否为素数
def sushu():
    i = int(input("请输入一个整数"))
    flag = False
    for x in range(2, i):
        if i % x == 0: 
            flag = True
            break

    if flag:
        print("是合数")
    else:
        print("素数")
```

```python
# 两直角边，求斜边
def xiebian1():
    l1 = float(input("输入直角边1"))
    l2 = float(input("输入直角边2"))
    l3 = (l1 ** 2 + l2 ** 2) ** 0.5
    print(l3)

def xiebian2():
    import math
    l1 = float(input("输入直角边1"))
    l2 = float(input("输入直角边2"))
    l3 = math.sqrt(l1 ** 2 + l2 ** 2)
    print(l3)

xiebian2()
```

```python
# 组合数字，1、2、3、4能组成多少个不同的无重复的三位数
count = 0
for a in range(1,5):
    for b in range (1,5):
        if a == b:
            continue
        else:
            for c in range(1,5):
                if c == a:
                    continue
                elif c == b:
                    continue
                else:
                    count += 1
                    print(f"{a}{b}{c}")

print(count)

for i in range(1,5):
    for j in range(1,5):
        for m in range(1,5):
            if (i != j) and (i != m) and (j != m):
                print(f"{i}{j}{m}")

```
