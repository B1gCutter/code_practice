```python
# 31.判断回文数
string = input("请输入一个数字，程序会判断其是否为回文数")
rev_string = string[::-1]
if rev_string == string:
    print("该数字是回文数")
else:
    print("该数字不是回文数")
```

```python
# 32.判断星期几
string = input("请输入日期")
if string[0] == "M":
    print("星期一")

elif string[0] == "T":
    if string[1] == "u":
        print("星期二")
    else:
        print("星期四")

elif string[0] == "W":
    print("星期三")

elif string[0] == "F":
    print("星期五")

elif string[0] == "S":
    if string[1] == "a":
        print("星期六")
    else:
        print("星期日")
```

```python
# 33.相反顺序输出列表
#  方法1
list = [1, 3, 2, 4, 5]
list.reverse()
print(list)
# 方法2
list1 = list[::-1]  #在reverse后的基础上再反向输出
print(list1)
```

```python
# 34.对十个数进行排序
num = map(int,input().split(","))   # map把int函数映射到列表内的每个元素上，得到迭代器
num = list(num)
num.sort()  # sort方法，把列表升序排列
print(num)
```

```python
# 35.求矩阵主对角线元素之和
a = []
sum = 0
for i in range(3):                  
    a.append([])
    for j in range(3):              # 内层作为矩阵的行，循环三次
        k = int(input("请输入数据"))  
        a[i].append(k)              # 输入的k传入矩阵a的第i行
        if i == j:
            sum += a[i][j]

print(sum)
```
