```python
a = int(input())        # 输入一个自然数，输出所有质数因子
y = 2                   
list = []               # 创建一个空列表

while a != y:           
    if a % y == 0:      
        list.append(y)  # 被除数和y不相等时，如果y为a的因数，导入到一个列表中
        a /= y          # 新自然数变为被因数除去之后的自然数，继续循环
    else:
        y += 1          # y被除不尽时+1，继续循环
list.append(int(a))     # 最后剩的a也是因数
print(list)


```

```python
# 找斐波那契数列中的第n项
num_position = int(input("要查找斐波那契数列的第几项呢？"))
list1 = [1,1]

for i in range(1,num_position - 1):
    if i != num_position:
        new_num = list1[-1] + list1[-2]
        list1.append(new_num)

print(list1)
print(f"第{num_position}项是{list1[-1]}")
```

```python
# 判断输入内容有多少字母，数字，以及其他字符
letter = 0
num = 0
char = 0
space = 0
sentence = input("请输入一段文字")
for l in sentence:
    if l.isalpha():
        letter += 1
    elif l.isdigit():
        num += 1
    elif l.isspace():
        space += 1
    else:
        char += 1

print(f"这段话一共有{letter}个字母{num}个数字{space}个空格{char}个其他字符")
```

```python
# 自由落体反弹
heigth = 100
distance = 100
for i in range(1,11):
    bounce = 1/2 ** i * heigth
    distance = 2 * bounce
    print(f"第{i}次落地时共经过{distance},弹{bounce}米")
```

```python
# 求指定数列的和
up = 2
down = 1
sum = 0

for i in range(20):
    sum += up/down
    a = down
    down = up
    up = up + a

print(sum)
```

```python
# 求各个阶乘的和
import math

sum = 0

for i in range(1,21):
    sum += math.factorial(i)

print(sum)



```
